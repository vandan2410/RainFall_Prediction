```python
df[["SUBDIVISION","ANNUAL"]].groupby("SUBDIVISION").sum().sort_values(by='ANNUAL',ascending=False).plot(kind='barh',stacked=True,figsize=(16,10))
plt.xlabel("Rainfall in MM",size=12)
plt.ylabel("Sub-Division",size=12)
plt.title("Annual Rainfall v/s SubDivisions")
plt.grid(axis="x",linestyle="-.")
plt.show()
```

```python
plt.figure(figsize=(15,8))
df.groupby("YEAR").sum()['ANNUAL'].plot(kind="line",color="r",marker=".")
plt.xlabel("YEARS",size=12)
plt.ylabel("RAINFALL IN MM",size=12)
plt.grid(axis="both",linestyle="-.")
plt.title("Rainfall over Years")
plt.show()
```

```python
df[['SUBDIVISION', 'JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL',
       'AUG', 'SEP', 'OCT', 'NOV', 'DEC']].groupby("SUBDIVISION").sum().plot(kind="barh",stacked=True,figsize=(13,8))
plt.title("Sub-Division v/s Rainfall in each month")
plt.xlabel("Rainfall in MM",size=12)
plt.ylabel("Sub-Division",size=12)
plt.grid(axis="x",linestyle="-.")
plt.show()
```

```python
vidharbhaRainfall= df.loc[((df['SUBDIVISION'] == 'VIDARBHA'))]
vidharbhaRainfall.head(5)
```

```python
X=vidharbhaRainfall.drop(['SUBDIVISION','YEAR','JAN', 'FEB','MAR','APR','MAY','JUN','OCT','NOV','DEC','ANNUAL','Jan-Feb','Mar-May','Jun-Sep','Oct-Dec'], axis=1)
```

```python
y=vidharbhaRainfall['ANNUAL']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
test_predictions=model.predict(X_test)
```

```python
test_predictions   #predicts y test
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
df['ANNUAL'].mean()
```

```python
mean_absolute_error(y_test,test_predictions)
```

```python
np.sqrt(mean_squared_error(y_test,test_predictions))
```
